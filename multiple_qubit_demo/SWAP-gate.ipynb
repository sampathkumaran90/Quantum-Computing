```python
from qiskit import QuantumCircuit, Aer, execute
from qiskit.visualization import plot_bloch_multivector, plot_histogram
from qiskit.visualization.array import array_to_latex
from qiskit.quantum_info import Operator, Statevector
```

```python
qc = QuantumCircuit(2)
qc.swap(0, 1) # Swaps the state of qubits 
qc.draw(output="mpl")
```

```python
backend = Aer.get_backend("statevector_simulator")
result = execute(qc, backend).result().get_statevector()
plot_bloch_multivector(result)
```

```python
result = execute(qc, backend).result().get_counts()
plot_histogram(result)
```

```python
backend = Aer.get_backend("unitary_simulator")
result = execute(qc, backend).result().get_unitary()
print(result)
```

```python
array_to_latex(Operator(qc))
```

```python
Statevector.from_label("00").evolve(qc).draw('latex')
```
